* frame_score stores engagement score per frame
* full stores maximum possible engagement store per frame
* final_score is calculated by taking mean over all frames
* Labels are generated by running YOLOv5 model over test images 

<font size="5">

```python
import pandas as pd
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
# assign directory
directory = 'D:\DATA\project\yolov5\\runs\detect\exp12\labels'
 
# iterate over files in
# that directory
files = Path(directory).glob('*')
files=sorted(files)
a=len(files)
all_data={}
final_score=0
thresh=0.3
frame=[]
frame_score=[]
full=[]
for fil in files:
  data = pd.read_csv(fil,sep=' ',names=['class','xmax','ymax','xmin','ymin','probability'])
  data= data.sort_values(by=['xmax'], ascending=True)
  col=data['probability']
  l=len(col)
  val=0
  k=0
  if(data['xmax'][k]<=thresh):
    if(data['class'][k]==1):
      if(data['probability'][k]>=0.5):
          val=val+(1-data['probability'][k])
      else:
        val=val+data['probability'][k]
    else:
      val=val+data['probability'][k]
  
  k=k+1
  val2=0
  if(data['xmax'][k]<=thresh):
    l=l-1
    if(data['class'][k]==1):
      if(data['probability'][k]>=0.5):
          val2=val2+(1-data['probability'][k])
      else:
        val2=val2+data['probability'][k]
    else:
      val2=val2+data['probability'][k]
  

  if(val2>val):
    k=k+1
    val=val2

  full.append(l)


  for i in range(k,len(col)):
    maxi=0
    if(data['xmax'][i]<=thresh):
      maxi=max(maxi,data['xmax'][i])
    else:
      if(data['class'][i]==1):
        val=val+data['probability'][i]
      else:
        if(data['probability'][i]>=0.5):
          val=val+(1-data['probability'][i])
        else:
          val=val+data['probability'][i]

    i=i+1
  frame.append(val)
  frame_score.append(val/l)

final_score=np.sum(frame)/np.sum(full)
print("Final score of Class engagement for this class is: ",final_score)
```
